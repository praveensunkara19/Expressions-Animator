```python
import os
import gradio as gr
import subprocess

# ============== Setup function called only once ==============
def setup():
    # Clone LivePortrait repo
    if not os.path.exists("LivePortrait"):
        os.system("git clone -b dev https://github.com/camenduru/LivePortrait")
    # Clone pretrained weights
    if not os.path.exists("LivePortrait/pretrained_weights"):
        os.system("git clone https://huggingface.co/camenduru/LivePortrait LivePortrait/pretrained_weights")

    # Build Cython extensions if not built
    cython_so = "LivePortrait/src/utils/dependencies/insightface/thirdparty/face3d/mesh/cython/mesh_core_cython.cpython-38-x86_64-linux-gnu.so"
    if not os.path.exists(cython_so):
        os.chdir("LivePortrait/src/utils/dependencies/insightface/thirdparty/face3d/mesh/cython")
        subprocess.run(["python3", "setup.py", "build_ext", "--inplace"])
        os.chdir("../../../../../../../../../")

setup()

# ============== Inference function ==============
def animate_face(source_img, driving_vid):
    if source_img is None or driving_vid is None:
        return "Please upload both source image and driving video."

    # Save inputs
    source_path = "source.jpg"
    output_path = "output.mp4"

    source_img.save(source_path)

    # driving_vid is a filepath from gr.Video input
    driving_path = driving_vid

    # Run LivePortrait inference
    cmd = f"python3 LivePortrait/inference.py -s {source_path} -d {driving_path} -o {output_path}"
    print(f"Running command: {cmd}")  # Debug print
    result = subprocess.run(cmd, shell=True)

    if result.returncode != 0 or not os.path.exists(output_path):
        return "Error during LivePortrait inference."

    return output_path

# ============== Gradio UI ==============
def create_interface():
    with gr.Blocks(theme=gr.themes.Soft()) as app:
        gr.Markdown("# 🎭 LivePortrait Face Reenactment")
        gr.Markdown(
            "Upload a **source image** and a **driving video (mp4)** to generate the animated output. "
            "Uses LivePortrait pretrained models for realistic reenactment."
        )

        with gr.Row():
            source_input = gr.Image(label="Source Image", type="pil")
            driving_input = gr.Video(label="Driving Video (mp4 only)")

        submit_btn = gr.Button("Generate Animation")
        output_video = gr.Video(label="Animated Output")

        submit_btn.click(animate_face, inputs=[source_input, driving_input], outputs=output_video)

        gr.Markdown("Developed by [Your Name or Lab]. Powered by LivePortrait + Gradio.")

    return app

if __name__ == "__main__":
    iface = create_interface()
    iface.launch(share=True, debug=True)  # share=True for Colab, remove for HF Spaces
```
